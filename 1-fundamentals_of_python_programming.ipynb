## 1.4 条件语句

条件要么是True要么是False，不会有别的判断结果。

需要注意：
1. Non-Zero = True
2. Non-Null = True

而：
1. Zero = False
2. Null = False

这是一个图示：



```python
age = int(input("Please enter the age: "))
if age < 18:
    print("未成年")
else:
    print("已经成年")
```

```python
age = int(input("Please enter the age: "))
if age < 18:
    print("未成年")
elif age >= 18 and age <= 60:
    print("成年")
else:
    print("老年")
```

## 1.5 循环

循环能够让我们不断的执行某一段代码：



但是我们需要有一个方法来跳出这个循环，避免陷入死循环的局面。

Python语言提供了两种循环方法：
1. while
2. for

这是一个图示：



```python
num = 10
while num <= 20:
    print(num)
    num += 1
```

```python
for char in "Hello World":
    print(char)
```

Python关于循环有三个有用的关键字：
1. break
2. continue
3. pass

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
num = 0
while num < 10:
    if num == 8:
        num += 1
        pass  # 跳过打印
    print(num)
    num += 1
```

```python
num = 0
while num < 10:
    if num == 6:
        num += 1
        continue
    print(num)
    num += 1
```
