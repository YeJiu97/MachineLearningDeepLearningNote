## 4.2 Turning Machine and Turing Test

图灵机：

>一条无限长的纸带TAPE。纸带被划分为一个接一个的小格子，每个格子上包含一个来自有限字母表的符号，字母表中有一个特殊的符号 口 表示空白。纸带上的格子从左到右依次被编号为0, 1, 2, ...，纸带的右端可以无限伸展。

> 一个读写头HEAD。它可以在纸带上左右移动，能读出当前所指的格子上的符号，并能改变它。

> 一套控制规则TABLE。它根据当前机器所处的状态以及当前读写头所指的格子上的符号来确定读写头下一步的动作，并改变状态寄存器的值，令机器进入一个新的状态，按照以下顺序告知图灵机命令：
1.写入（替换）或擦除当前符号；
2.移动 HEAD， 'L'向左， 'R'向右或者'N'不移动；
3.保持当前状态或者转到另一状态。

> 一个状态寄存器。它用来保存图灵机当前所处的状态。图灵机的所有可能状态的数目是有限的，并且有一个特殊的状态，称为停机状态。参见停机问题。

## 4.3 What Is Deep Learning

深度学习是机器学习的一个子领域，或者说是一种方法。

人工智能，机器学习和深度学习之间的关系如下所示：



几种学习：
1. Supervised：有监督学习，给定了输入和输出，让机器去寻找数据之间的规律
2. Unsupervised：无监督学习，只给出了输入数据，让机器去寻找数据之间的关系，或者说进行分类
3. Reinforcement Learning：强化学习，强化某一个/一系列行为让模型去寻找出在某一个环境之下的最大期望的行为

哪些会导致糟糕的结果：
1. 没有足够的数据
2. 计算性能落后
3. 错误的初始化
4. 不正确的非线性激活函数

深度学习：

1. 深度学习可以在没有监督的情况之下进行，深度学习是一种五监督学习
2. 深度学习需要大量的数据，数据量越大，训练的结果越准确
3. 需要多层，multiple layers

## 4.5 Machine Learning Method

方法一：概率建模 Probabilistic Model

方法二：Bayes‘s Therorem

方法三：逻辑回归 Logistic Regression

方法四：分类算法 Classification Algorithm

Kernel Methods：偏差-方差困境在机器学习方法中占主导地位。如果模型太简单，模型将难以找到输入和输出之间的适当关系。然而，如果一个模型太复杂，它在训练中会表现得更好，但在看不见的数据上会经历更大的性能差异，而且复杂的模型通常计算成本更高。在理想的世界中，我们需要一个简单的模型，它可以快速训练，并且足够复杂，可以找到输入和输出之间的复杂关系。内核方法通过将数据的输入空间映射到更高维的特征空间来做到这一点，在其中可以训练简单的线性模型，从而产生高效、低偏差、低方差的模型。

Support Vector Machines（SVM）：二分类模型，它的基本模型是定义在特征空间上的间隔最大的线性分类器，间隔最大使它有别于感知机。

## 4.6 Supervised Machine Learning Method

**线性回归Logistic Regression**

线性回归的核心逻辑：最小化模型的误差/做出最准确的预测。



三个组成部分：
1. 一个模型
2. 一组输入数值
3. 一组输出结果对应输入的数值
4. 线性方程为每一个输入值分配一个比例因子，也就是系数和，使用 β 来进行表示
5. 一个额外的系数

系数会影响直线的坡度，而额外系数则会让直线进行上下移动。



以上是线性回归的逻辑，而逻辑回归则是再次之上进行了借鉴。

逻辑回归实际上是一种go to的二分法，其核心函数被称之为叫做逻辑函数（Logistic function）。



结果要么是0要么是1。

**决策书 Decision Tree**

决策树图示如下所示：



决策树的优点：容易理解，可以进行可视化分析，能够处理不相关的特征。

决策书的劣势：容易发生过拟合，容易忽视数据集中属性的相关性。

随机森林是集成算法中的bagging方法，是多个决策树组合器。每个决策树不同，提高系统的多样性，提升分类性能：
1. 对训练数据重复有放回抽样，抽样样本为训练数据样本量
2. 对特征随机抽样
3. 对每个抽样样本，根据选择的特征进行学习决策树
4. 由多个独立的决策树进行多数表决得到分类结果

**朴素贝叶斯 Naive Bayes**

什么是贝叶斯定理：



对于这个公式而言：

1. P(A)：表示A出现的概率
2. P(B|A) 是条件概率的符号，表示在事件A发生的条件之下，事件B发生的概率，P(B|A)被称之为叫做似然度
3. P(A|B)也是条件概率的符号，表示在事件B发生的概率之下，事件A发生的可能性，这个计算的结果被称之为叫做后验概率

贝叶斯公式可以预测事件发生的概率，两个本来相互独立的事件，发生了某种“相关性”，此时就可以通过“贝叶斯公式”实现预测。

机器学习中是一系列以假设特征之间强（朴素）独立下运用贝叶斯定理为基础的简单概率分类器

分类器模型会给问题实例分配用特征值表示的类标签，类标签取自有限集合。它不是训练这种分类器的单一算法，而是一系列基于相同原理的算法：所有单纯贝氏分类器都假定样本每个特征与其他特征都不相关。
