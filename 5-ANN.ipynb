## 5.1 What is Artificial Neural Network

在机器学习和认知科学领域，是一种模仿生物神经网络（动物的中枢神经系统，特别是大脑）的结构和功能的数学模型或计算模型，用于对函数进行估计或近似。

神经网络由大量的人工神经元联结进行计算。大多数情况下人工神经网络能在外界信息的基础上改变内部结构，是一种自适应系统，通俗地讲就是具备学习功能。

## 5.2 Anatomy of Neural Network

Neural Network有这样的组成：

1. Layer
2. input data and corresponding targets
3. loss function
4. optimizer



## 5.3 Keras API
